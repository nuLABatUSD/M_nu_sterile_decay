```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from classy import Class
import Neutrino_Work as NW
```

```python
small_m = np.logspace(-3,-1,100)
neutrino_masses_t = np.zeros(len(small_m))
for i in range(len(small_m)):
    neutrino_masses_t[i] = NW.v_masses(small_m[i],True)[0] + NW.v_masses(small_m[i],True)[1] + NW.v_masses(small_m[i],True)[2]
neutrino_masses_f = np.zeros(len(small_m))
for i in range(len(small_m)):
    neutrino_masses_f[i] = NW.v_masses(small_m[i],False)[0] + NW.v_masses(small_m[i],False)[1] + NW.v_masses(small_m[i],False)[2]

    
    
plt.figure()
plt.loglog(small_m,neutrino_masses_t)
plt.loglog(small_m,neutrino_masses_f)
plt.xlabel('m_small')
plt.ylabel('m1+m2+m3')
plt.show()
```
